```python
#EJERCICIO 1: Historial de Atenciones - Lista Enlazada Simple
class AtencionMedica:
    """Nodo que representa una atención médica finalizada."""
    def __init__(self, id_registro, nombre_paciente, fecha, monto_consulta):
        self.id_registro = id_registro
        self.nombre_paciente = nombre_paciente
        self.fecha = fecha
        self.monto_consulta = monto_consulta
        self.sig = None 

class HistorialSimple:
    """Estructura de Lista Enlazada Simple."""
    def __init__(self):
        self.cabeza = None

    def insertar_registro(self, id_registro, nombre_paciente, fecha, monto_consulta):
        """Inserta un nuevo registro siempre al final de la lista."""
        nuevo_nodo = AtencionMedica(id_registro, nombre_paciente, fecha, monto_consulta)
        if not self.cabeza:
            self.cabeza = nuevo_nodo
            print(f"Registro {id_registro} insertado como cabeza del historial.")
            return
        
        actual = self.cabeza
        while actual.sig:
            actual = actual.sig
        actual.sig = nuevo_nodo
        print(f"Registro {id_registro} insertado al final del historial.")

    def mostrar_historial(self):
        """Recorrido simple para mostrar todas las atenciones."""
        if not self.cabeza:
            print("El historial de atenciones está vacío.")
            return
        
        print("\nHISTORIAL COMPLETO DE ATENCIONES")
        actual = self.cabeza
        while actual:
            print(f"ID: {actual.id_registro} | Paciente: {actual.nombre_paciente} | Fecha: {actual.fecha} | Monto: ${actual.monto_consulta}")
            actual = actual.sig

    def filtrar_por_monto(self, valor_minimo):
        """Muestra solo las consultas cuyo costo sea estrictamente mayor al valor ingresado."""
        if not self.cabeza:
            print("El historial está vacío. No hay montos que evaluar.")
            return
        
        print(f"\n Consultas con monto mayor a: ${valor_minimo}")
        actual = self.cabeza
        encontrado = False
        while actual:
            if actual.monto_consulta > valor_minimo:
                print(f"   [Filtrado] ID: {actual.id_registro} -> Paciente: {actual.nombre_paciente} | Monto: ${actual.monto_consulta}")
                encontrado = True
            actual = actual.sig
        
        if not encontrado:
            print("No se encontraron consultas que superen el monto ingresado.")
```

```python
historial = HistorialSimple()

print("PRUEBA DE INSERCIÓN")
historial.insertar_registro(1, "Daniel Pomboza", "18/05/2026", 45.0)
historial.insertar_registro(2, "Saul Masapanta", "18/05/2026", 120.0)
historial.insertar_registro(3, "Daniel Carillo", "19/05/2026", 25.0)

historial.mostrar_historial()

print("\nPRUEBA DE FILTRADO")
monto_limite = float(input("Ingrese el monto mínimo para filtrar las consultas: "))
historial.filtrar_por_monto(monto_limite)
```

```python
#EJERCICIO 2: Sala de Espera
class TurnoEspera:
    """Nodo para la sala de espera con conectividad doble."""
    def __init__(self, numero_turno, nombre_paciente, hora_llegada, motivo_consulta):
        self.numero_turno = numero_turno
        self.nombre_paciente = nombre_paciente
        self.hora_llegada = hora_llegada
        self.motivo_consulta = motivo_consulta
        self.sig = None
        self.ant = None

class SalaEsperaDoble:
    """Estructura de Lista Doble Enlazada."""
    def __init__(self):
        self.cabeza = None

    def llegada_paciente(self, numero_turno, nombre_paciente, hora_llegada, motivo_consulta):
        """Inserta un nuevo turno al final de la lista de espera."""
        nuevo_turno = TurnoEspera(numero_turno, nombre_paciente, hora_llegada, motivo_consulta)
        if not self.cabeza:
            self.cabeza = nuevo_turno
            print(f"✔ Turno {numero_turno} asignado al inicio de la sala.")
            return
        
        actual = self.cabeza
        while actual.sig:
            actual = actual.sig
        
        actual.sig = nuevo_turno
        nuevo_turno.ant = actual
        print(f"✔ Turno {numero_turno} enviado al final de la cola de espera.")

    def navegar_paciente(self, numero_turno):
        """Busca un paciente y muestra explícitamente quién está antes y después."""
        actual = self.cabeza
        while actual:
            if actual.numero_turno == numero_turno:
                print(f"\nNavegación desde el Turno {numero_turno} ({actual.nombre_paciente})")
                
                if actual.ant:
                    print(f"Anterior: Turno {actual.ant.numero_turno} - {actual.ant.nombre_paciente}")
                else:
                    print("Anterior: [No hay nadie antes, es el primero en la cola]")
                
                if actual.sig:
                    print(f"Siguiente: Turno {actual.sig.numero_turno} - {actual.sig.nombre_paciente}")
                else:
                    print(" Siguiente: [No hay nadie más, es el último de la cola]")
                return
            actual = actual.sig
        print(f"El turno {numero_turno} no se encuentra en la sala de espera.")

    def prioridad_emergencia(self, numero_turno):
        """Mueve un turno específico directamente al inicio de la lista modificando enlaces."""
        if not self.cabeza:
            print("La sala está vacía.")
            return

        # Si ya es la cabeza, no requiere movimiento
        if self.cabeza.numero_turno == numero_turno:
            print(f"El turno {numero_turno} ya se encuentra al inicio de la prioridad.")
            return

        actual = self.cabeza
        while actual:
            if actual.numero_turno == numero_turno:
                # 1. Desconectar el nodo de su posición actual
                if actual.sig:
                    actual.sig.ant = actual.ant
                actual.ant.sig = actual.sig

                # 2. Insertarlo al inicio (como nueva cabeza)
                actual.sig = self.cabeza
                actual.ant = None
                self.cabeza.ant = actual
                self.cabeza = actual
                print(f"EMERGENCIA: El Turno {numero_turno} ({actual.nombre_paciente}) ha sido movido al INICIO.")
                return
            actual = actual.sig
        print(f"No se encontró el turno {numero_turno} para aplicar prioridad.")

    def mostrar_sala(self):
        """Muestra el estado actual de la fila."""
        if not self.cabeza:
            print("Sala de espera vacía.")
            return
        print("\nFILA ACTUAL EN SALA DE ESPERA:")
        actual = self.cabeza
        while actual:
            print(f"[{actual.numero_turno} | {actual.nombre_paciente}]", end=" <=> ")
            actual = actual.sig
        print("Fin de la cola")
```

```python
sala = SalaEsperaDoble()
sala.llegada_paciente(101, "Paciente Alfa", "08:00", "Control")
sala.llegada_paciente(102, "Paciente Beta", "08:15", "Dolor estomacal")
sala.llegada_paciente(103, "Paciente Gamma", "08:20", "Fractura")

sala.mostrar_sala()
sala.navegar_paciente(102)

sala.prioridad_emergencia(103)
sala.mostrar_sala()
```

```python
class MonitoreoPaciente:
    """Nodo clínico para monitoreo continuo en UCI."""
    def __init__(self, id_cama, nombre_paciente, frecuencia_cardiaca, estado_alerta):
        self.id_cama = id_cama
        self.nombre_paciente = nombre_paciente
        self.frecuencia_cardiaca = frecuencia_cardiaca
        self.estado_alerta = estado_alerta
        self.sig = None
        self.ant = None

class MonitorUciCircular:
    """Estructura de Lista Doble Circular."""
    def __init__(self):
        self.cabeza = None

    def conectar_paciente(self, id_cama, nombre_paciente, frecuencia_cardiaca, estado_alerta):
        """Añade un paciente al anillo de monitoreo continuo."""
        nuevo_paciente = MonitoreoPaciente(id_cama, nombre_paciente, frecuencia_cardiaca, estado_alerta)
        
        if not self.cabeza:
            self.cabeza = nuevo_paciente
            nuevo_paciente.sig = nuevo_paciente
            nuevo_paciente.ant = nuevo_paciente
            print(f"Cama {id_cama}: Primer paciente conectado al monitor central.")
            return

        ultimo = self.cabeza.ant
        
        ultimo.sig = nuevo_paciente
        nuevo_paciente.ant = ultimo
        nuevo_paciente.sig = self.cabeza
        self.cabeza.ant = nuevo_paciente
        print(f" Cama {id_cama}: Paciente integrado exitosamente al flujo cíclico.")

    def desconectar_paciente(self, id_cama):
        """Elimina un paciente del ciclo y restaura la continuidad del anillo."""
        if not self.cabeza:
            print("No hay pacientes conectados al monitor.")
            return

        actual = self.cabeza
        encontrado = False

        while True:
            if actual.id_cama == id_cama:
                encontrado = True
                break
            actual = actual.sig
            if actual == self.cabeza:
                break

        if not encontrado:
            print(f"No se encontró la cama {id_cama} en el sistema de monitoreo.")
            return
        
        if actual.sig == actual:
            self.cabeza = None
            print(f"Cama {id_cama} desconectada. Monitor sin pacientes.")
            return

        actual.ant.sig = actual.sig
        actual.sig.ant = actual.ant
        if actual == self.cabeza:
            self.cabeza = actual.sig

        print(f" Cama {id_cama} desconectada con éxito. Ciclo reestructurado.")

    # PARTE PERSONALIZADA: Simulación Cíclica Activa con Telemetría de Alertas
    def simular_ciclo_monitoreo_con_alertas(self, vueltas_totales):
        """
        Simula el escaneo del monitor pasando de un paciente a otro de forma infinita/cíclica.
        Si detecta un paciente en estado 'Crítico' o arritmias, lanza alertas de prioridad UCI.
        """
        if not self.cabeza:
            print("El monitor no registra pacientes activos.")
            return

        print(f"\nINICIANDO SIMULACIÓN DE ESCANEO CENTRAL UCI ({vueltas_totales} vueltas completas)")
        actual = self.cabeza
        nodos_a_recorrer = 0
        
        temp = self.cabeza
        while True:
            nodos_a_recorrer += 1
            temp = temp.sig
            if temp == self.cabeza:
                break
        
        iteraciones_totales = nodos_a_recorrer * vueltas_totales
        
        for i in range(iteraciones_totales):
            print(f"\n[Escaneando...] Mostrando Pantalla de Cama {actual.id_cama}")
            print(f" -> Paciente: {actual.nombre_paciente} | FC: {actual.frecuencia_cardiaca} lpm | Estado: {actual.estado_alerta}")
            
            if actual.estado_alerta.lower() == "crítico" or actual.frecuencia_cardiaca > 120 or actual.frecuencia_cardiaca < 50:
                print(f"ALERT TRIGGERED en Cama {actual.id_cama}!! Asistencia inmediata requerida.")
            
            actual = actual.sig 
```

```python
monitor = MonitorUciCircular()
monitor.conectar_paciente(10, "Paciente A", 72, "Estable")
monitor.conectar_paciente(11, "Paciente B (Crítico)", 135, "Crítico") 
monitor.conectar_paciente(12, "Paciente C", 64, "En Observación")

monitor.simular_ciclo_monitoreo_con_alertas(vueltas_totales=2)

monitor.desconectar_paciente(11)
monitor.simular_ciclo_monitoreo_con_alertas(vueltas_totales=1)
```
